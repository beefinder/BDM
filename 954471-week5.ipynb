# Ex. 1 Please perform preprocessing 

```python
df = pd.read_csv('https://raw.githubusercontent.com/plenoi/Clinic/master/ultima_all_clean.csv', sep=',', header = 0) #บอกหัวตารางด้วย ถ้าไม่ีจะใส่ none
df
```

**1. Set index of data to hn numbers**

```python
df_set = df
df_set = df_set.set_index('hn') 
df_set
```

**2. Remove missing values**

```python
df_set.isnull()
```

```python
df_clean = df_set.dropna(0, 'any') 
df_clean
```

**3. Remove these unnecessary columns hiv, size, utmet, vgmet, surgery, pchemo**

```python
df_clean.columns
```

```python
df_clean = df_clean.drop(['hiv', 'size', 'utmet', 'vgmet', 'surgery', 'pchemo'],1)
df_clean.columns
```

**4. Change age into 2 values 'Young' (age <= 20) and 'Old' (age > 20)**

```python
df_age = df_clean
df_age.loc[df_age['age']<=20, 'age'] = 'Young'
df_age.loc[df_age['age']>20, 'age'] = 'Old'
df_age
```

**5.Separate data into groups based on pmmet values**

**pmmet = 0**

```python
pmmet0 = df_age.loc[df_age['pmmet']==0.0]
pmmet0
```

**pmmet = 1**

```python
pmmet1 = df_age.loc[df_age['pmmet']==1.0]
pmmet1
```

**6. for each group, randomly choose 10 data**

```python
pmmet0.sample(10)
```

```python
pmmet1.sample(10)
```
