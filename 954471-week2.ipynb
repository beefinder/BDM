```python
def power(a, b):
    c=a**b
    return c

power(2,3)
```

```python
def summation(L):
    z = 0
    for item in L:
        z=z+item
    return z

summation([1,2,3,4,5])
```

```python
def check(number):
    even = len([i for i in number if i%2==0])
    return('Number of even numbers is '+str(even)+'\n'+'Number of odd numbers is '+str(len(number)-even))
        
print(check([1,2,3,4,5,6,7,8,9]))
```
